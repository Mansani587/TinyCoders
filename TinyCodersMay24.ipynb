# 18. Add column 'Age' in Patient table

```python
from datetime import date

# Calculating Age using the function
def calculateAge(DateOfBirth):
    today = date.today()
    age = today.year - DateOfBirth.year - ((today.month, today.day) <
         (DateOfBirth.month, DateOfBirth.day))
    return age

df = pd.DataFrame(PA)
df2 = pd.DataFrame().assign(DateOfBirth=pd.to_datetime(df.DateOfBirth))
df2['age']=df2['DateOfBirth'].apply(lambda x: calculateAge(x))
df['age'] = df2['age']
df
```

# 12. Connect to sql and Write a query to get list of Provider names whose Providername is starting with letter T

```python
pip install psycopg2
```

```python
import psycopg2
import psycopg2.extras

hostname = 'localhost'
database = 'HospitalDatabase'
username = 'postgres'
pwd = 'postgres'
port_id = 5432
conn = None

try:
    with psycopg2.connect(
                host = hostname,
                dbname = database,
                user = username,
                password = pwd,
                port = port_id) as conn:

        with conn.cursor(cursor_factory=psycopg2.extras.DictCursor) as cur:

            cur.execute("""select * from "Providers" where "ProviderName" like 'T%'""")
            for record in cur.fetchall():
                
                print(record["Provider_ID"],record["ProviderName"],record["PS_ID"],record["ProviderDateOnStaff"])
except Exception as error:
    print(error)
finally:
    if conn is not None:
        conn.close()
```
