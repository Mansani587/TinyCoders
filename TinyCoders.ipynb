18. Add column 'Age' in Patient table

```python
from datetime import date

# Calculating Age using the function
def calculateAge(DateOfBirth):
    today = date.today()
    age = today.year - DateOfBirth.year - ((today.month, today.day) <
         (DateOfBirth.month, DateOfBirth.day))
    return age

df = pd.DataFrame(PA)
df2 = pd.DataFrame().assign(DateOfBirth=pd.to_datetime(df.DateOfBirth))
df2['age']=df2['DateOfBirth'].apply(lambda x: calculateAge(x))
df['age'] = df2['age']
df
```

3. "Create a figure object called fig using plt.figure() **
** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
** Plot (x,y) on that axes and set the labels and titles to match the plot below:**
Note: You can plot on any x,y set of values"

```python
##### %matplotlib inline
x = range(5, 10)
y = range(5, 10)

# creating a figure object

fig1=plt.figure()

New_axes=fig1.add_axes([0,0,1,1])

New_axes.plot(x,y)

New_axes.set_xlabel("x-axis",fontsize=20)

New_axes.set_ylabel("y-axis",fontsize=20)
#axs.set_title('Negative correlation')
New_axes.set_title("Line Graph",fontsize=20)
#plt.show()
```
